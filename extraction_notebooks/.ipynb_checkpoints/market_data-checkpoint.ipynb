```python
from splinter import Browser
import requests
from webdriver_manager.chrome import ChromeDriverManager
from pycoingecko import CoinGeckoAPI
from datetime import datetime
import pandas as pd

cg = CoinGeckoAPI()
```

```python
#splinter test browser
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
#url for request of dogecoin price data for max days at hourly interval
url = 'https://api.coingecko.com/api/v3/coins/dogecoin/market_chart?vs_currency=usd&days=max&interval=hourly'
browser.visit(url)
```

Get historical market data include price, market cap, and 24h volume within a range of timestamp (granularity auto)

Data granularity is automatic (cannot be adjusted)
1 day from query time = 5 minute interval data
1 - 90 days from query time = hourly data
above 90 days from query time = daily data (00:00 UTC)

```python
max_data = cg.get_coin_market_chart_by_id(id='dogecoin', vs_currency='usd', days='max', interval='hourly')
max_data
```

```python
max_df = pd.DataFrame(max_data)
```

max_df

```python
max_df
```

price_df

```python
#includes all params
price_df = pd.DataFrame(max_df['prices'].to_list(),columns=['timestamp','price'])
price_df
```

```python
market_df = pd.DataFrame(max_df['market_caps'].to_list(),columns=['timestamp','market_cap'])
market_df
```

```python
volume_df = pd.DataFrame(max_df['total_volumes'].to_list(),columns=['timestamp','volume'])
volume_df
```

```python
from functools import reduce

combined = [price_df, market_df, volume_df]

combined_df = reduce(lambda left,right: pd.merge(left,right,on='timestamp'), combined)
combined_df
```

```python
dates = []
for date in combined_df['timestamp']:
    datetime = datetime.utcfromtimestamp(date //1000)
    dates.append(datetime)

combined_df['date'] = dates
combined_df.sort_values(by='date')
```
